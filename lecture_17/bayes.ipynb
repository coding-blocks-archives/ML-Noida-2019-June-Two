```python
class NB:
    
    def fit(self, X, y):
        model = {}
        prior = {}

        for klass in set(y):
            model[klass] = {}
            selcted = X[y==klass]
            prior[klass] = len(selcted)/len(X)
            for column in X.columns:
                model[klass][column] = {}
                for value in set(X[column]):
                    model[klass][column][value] = np.sum(selcted[column] == value)/len(selcted)
        
        self.model = model
        self.prior = prior
        
    
    def predict_point(self, point):
        
        probs = []
        for klass in self.model:
            p = self.prior[klass]
            for column in self.model[klass]:
                value = point[column]
                p *= self.model[klass][column][value]
            
            probs.append(p)
        
        return np.argmax(probs)
    
    def predict(self, X):
        y = []
        for index, row in X.iterrows():
            y.append(self.predict_point(row))
        
        return np.array(y)
    
    def score(self, X, y):
        yp = self.predict(X)
        return np.mean(y == yp)
          
```
