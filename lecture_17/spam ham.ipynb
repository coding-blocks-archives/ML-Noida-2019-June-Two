```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
msgs = pd.read_csv("../bayes/spam.csv", encoding="ISO-8859-1")
```

```python
X = msgs['v2']
y = msgs['v1']
```

```python
msgs.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
st = set(stopwords.words("english"))
punc = set(punctuation)
ps = PorterStemmer()

def custom(sent):
    words = word_tokenize(sent.lower())
    faaltu = st.union(punc)
    pure = [w for w in words if w not in faaltu]
    more_pure = list(map(ps.stem, pure))
    
    return more_pure
```

```python
cv = CountVectorizer(tokenizer=custom)
```

```python
cv.fit(X)
```

```python
X_gen = cv.transform(X).todense()
```
